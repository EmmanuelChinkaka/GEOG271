```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Read dataset and display columns

```python
df=pd.read_csv("https://raw.githubusercontent.com/bot13956/ML_Model_for_Predicting_Ships_Crew_Size/master/cruise_ship_info.csv")
df
```

```python
df.head()
```

2. Calculate basic statistics of the data

```python
df.describe()
```

```python
cols = ['Age', 'Tonnage', 'passengers', 'length', 'cabins','passenger_density','crew']
cols
```

```python
sns.pairplot(df[cols], size=2.0)
```

**Observations from part 2**

We observe that variables are on different scales, for sample the Age variable ranges from about 16 years to 48 years, while the Tonnage variable ranges from 2 to 220, see probability density plots below. It is therefore important that when a regression model is built using these variables, variables be brought to same scale either by standardizing or normalizing the data.

We also observe that the target variable 'crew' correlates well with 4 predictor variables, namely, 'Tonnage', 'passengers', 'length', and 'cabins'

```python
dt = ['Age','Tonnage']
sns.pairplot(df[dt], size=2.0)
```
